```python
# display information about the dataset
def display_info(x):
    print("===================================================")
    print("Display the Infomation of DataSet")
    print()
    print(x.info())
    print("===================================================")
    print("Display the Columns of DataSet")
    print()
    print(x.columns)
    print("===================================================")
    print("Display the Stats of DataSet")
    print()
    print(x.describe())
    print("===================================================")
    print("Display the no of rows and columns of DataSet")
    print()
    row, cols = x.shape
    print(f"Rows = {row} Columns = {cols}")
    print("===================================================")
    print("Display the NUll Values of DataSet")
    print()
    print(x.isna().sum())
    print("===================================================")
    print("Display the Duplicates Values of DataSet")
    print()
    print(x.duplicated().sum())
```
